```python
import panel as pn
pn.extension()
import random
import matplotlib.pyplot as plt

#from dashLib.dash_tools import Dasht

import folium
import fiona
import geopandas as gpd

from dashLib.dora_tools import get_perf_data_summary, get_perf_df, kit_mgrs_count
```

```python
def add_my_aois(my_gdf, m):
    print("Shape",my_gdf.shape)
    for _, r in my_gdf.iterrows():
        # Without simplifying the representation of each borough,
        # the map might not be displayed
        sim_geo = gpd.GeoSeries(r['geometry']).simplify(tolerance=0.001)
        geo_j = sim_geo.to_json()
        geo_j = folium.GeoJson(data=geo_j,
                            style_function=lambda x: {'fillColor': 'orange'})
        folium.Popup(r['Name']).add_to(geo_j)
        geo_j.add_to(m)
        print(r['Name'])
    return m
```

```python
def organize_table():
    df=get_perf_df('kitware')
    mdf = kit_mgrs_count(df)
    mdf = mdf.head(18)
    df_widget = pn.widgets.DataFrame(mdf, show_index=True,
             fit_columns=True, row_height=23, name='Kit STuff')

    return pn.panel(df_widget)
    
```

```python
def get_plot_wrs2():
    df=get_perf_df('kitware')
    mdf = kit_mgrs_count(df)
    mdf = mdf.head(18)
    plot_o = pn.pane.Matplotlib(mdf.plot.barh().get_figure())

    return plot_o

def get_plot_mgrs():
    return '## MGRS PLOT HERE'
```

```python
def organize_plot(wrs2_or_mgrs):
    if wrs2_or_mgrs == 'WRS2':
        my_plot = get_plot_wrs2()
    else:
        my_plot = get_plot_mgrs()

    return my_plot
```

```python
df=get_perf_df('kitware')
mdf = kit_mgrs_count(df)
mdf.reset_index(inplace=True)
mdf
mgrs_tile_list = mdf['MGRS'].to_list()
```

```python
def organize_map():
    df=get_perf_df('kitware')
    mdf = kit_mgrs_count(df)
    mdf.reset_index(inplace=True)

    mgrs_tile_list = mdf['MGRS'].to_list()
    print(mgrs_tile_list)
    m = folium.Map(location=[0, 0], zoom_start=4, tiles='CartoDB positron')

    gpd.io.file.fiona.drvsupport.supported_drivers['KML'] = 'rw'
    mgrs_kml_gdf = gpd.read_file('/usmart/kml/S2A_OPER_GIP_TILPAR_MPC__20151209T095117_V20150622T000000_21000101T000000_B00.kml', driver='KML')
    my_gdf = mgrs_kml_gdf[mgrs_kml_gdf['Name'].isin(mgrs_tile_list)]
    add_my_aois(my_gdf, m)
    
    return m
```

```python
# title = '# Panel App \n- Where Does my Bucket Data Live Dora?'


title = '# Dora --      '
```

```python
def something_interesting(lsv, plotv):
    if plotv == 'Plot':
        return organize_plot(lsv)
    elif plotv == 'Map':
        return organize_map()
    else:
        return organize_table()
```

```python
static_text = pn.widgets.StaticText(name='EventSeed', value='A string')
static_text
```

```python
ls_radio = pn.widgets.RadioBoxGroup(name='radio1', options=['WRS2','MGRS'], inline=True)
ls_radio
```

```python
plot_radio = pn.widgets.RadioBoxGroup(name='radio2', options=['Plot','Table','Map'], inline=True)
plot_radio
```

```python
@pn.depends(ls_radio.param.value, plot_radio.param.value)
def duh(ls_val, plot_val):
    static_text.value = f'{ls_val} -- {plot_val} --'
    return something_interesting(ls_val,plot_val)
```

```python
#duh('a','Plot')
```

```python
app = pn.Column(pn.Row(title,  plot_radio),
        pn.Row(static_text, ls_radio),
        pn.Row(pn.panel(duh, height=600, width=800)))
  
```
