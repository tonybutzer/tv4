```python
import panel as pn
pn.extension()
```

```python
global cnt
cnt=1
# the key word 'save' can never be changed when the callback is triggered.
def update(event, save=True): 
    global cnt
    cnt = cnt +1 
    print(event)
    print(save)
    print('hello')
    widget.value = cnt
    return save # which variable should take and use the return value? The calling Button?

button = pn.widgets.Button(name='CH UP')
button.on_click(update)

buttond = pn.widgets.Button(name='CH Down')
buttond.on_click(update)

widget = pn.widgets.TextInput(name='Channel=', value='A string')

# using a bokeh server inside the notebook the print statements in the method 'update' are displayed,
# Otherwise the print statements are not visible in the outout cell
layout = pn.Row(button, buttond, widget)
layout.servable()
```

```python
def write_log(event):
    
    f = open("demofile2.txt", "a")
    f.write("See you soon!")
    f.close()
```

```python


backward = pn.widgets.Button(name='\u25c0', width=50)
forward = pn.widgets.Button(name='\u25b6', width=50)
search = pn.widgets.Button(name='🔍', width=100)

backward.on_click(write_log)
backward.clicks += 1


pn.Row(backward, forward, search).servable()
```

```python
text = pn.widgets.TextInput(value='Ready')

def b(event):
    text.value = 'Clicked {0} times'.format(button.clicks)
    
button.on_click(b)
button.clicks += 1
pn.Row(button, text)
#print(button.clicks)
```
