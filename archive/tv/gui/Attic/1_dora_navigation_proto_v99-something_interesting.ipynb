```python
import panel as pn
pn.extension()
import random

#from dashLib.dash_tools import Dasht

import folium
```

```python
title = '# Panel App \n- Where Does my Bucket Data Live Dora?'
```

```python
def something_interesting(lsv, plotv):
    if plotv == 'Plot':
        return title + lsv
    else:
        return random.randint(1,100)
```

```python
something_interesting('a','Plot')
```

```python
static_text = pn.widgets.StaticText(name='EventSeed', value='A string')
static_text
```

```python
ls_radio = pn.widgets.RadioBoxGroup(name='radio1', options=['WRS2','MGRS'], inline=True)
ls_radio
```

```python
plot_radio = pn.widgets.RadioBoxGroup(name='radio2', options=['Plot','Table'], inline=True)
plot_radio
```

```python
@pn.depends(ls_radio.param.value, plot_radio.param.value)
def duh(ls_val, plot_val):
    static_text.value = f'{ls_val} -- {plot_val}'
    return something_interesting(ls_val,plot_val)
```

```python
#duh('a','Plot')
```

```python
app = pn.Column(pn.Row(title, ls_radio, plot_radio),
        pn.Row(static_text),
        pn.Row(pn.panel(duh, height=600, width=800)))
  
```

```python
app.servable()
```
